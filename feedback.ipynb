```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of alpha current left values
alpha_left_current_values = np.linspace(0, 20, 100)
alpha_right_current_values = np.linspace(0, 20, 100)
# Define baseline values
alpha_left_baseline = 5
alpha_right_baseline = 10
def function1(alpha_left_current_values, alpha_right_current_values, alpha_left_baseline, alpha_right_baseline):
    return ((alpha_left_current_values - alpha_left_baseline) - (alpha_right_current_values - alpha_right_baseline)) / ((alpha_left_current_values - alpha_left_baseline) + (alpha_right_current_values - alpha_right_baseline))

def function2(alpha_left_current_values, alpha_right_current_values, alpha_left_baseline, alpha_right_baseline):
    return (alpha_left_current_values - alpha_right_current_values) / (alpha_left_current_values + alpha_right_current_values) - (alpha_left_baseline - alpha_right_baseline) / (alpha_left_baseline + alpha_right_baseline)
                                                                                                                                                                                 
# Plotting
fig, axes = plt.subplots(2, 2, figsize=(10, 6))

function1_values = function1(alpha_left_current_values, 5, alpha_left_baseline, alpha_right_baseline)
function2_values = function2(alpha_left_current_values, 5, alpha_left_baseline, alpha_right_baseline)
axes[0,0].plot(alpha_left_current_values, function1_values, label='Function 1')
axes[1,0].plot(alpha_left_current_values, function2_values, label='Function 2')

function1_values = function1(5, alpha_right_current_values, alpha_left_baseline, alpha_right_baseline)
function2_values = function2(5, alpha_right_current_values, alpha_left_baseline, alpha_right_baseline)
axes[0,1].plot(alpha_right_current_values, function1_values, label='Function 1')
axes[1,1].plot(alpha_right_current_values, function2_values, label='Function 2')


# Make a 3D plot function of alpha_right_current_values and alpha_left_current_values

# Plotting
```

```python
%matplotlib qt

from matplotlib import cm
from matplotlib.ticker import LinearLocator



import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


# Define the range of alpha current values


# Define baseline values
alpha_left_baseline = 5
alpha_right_baseline = 10

# Define functions
def function1(alpha_left_current, alpha_right_current, alpha_left_baseline, alpha_right_baseline):
    return ((alpha_left_current - alpha_left_baseline) - (alpha_right_current - alpha_right_baseline)) / ((alpha_left_current - alpha_left_baseline) + (alpha_right_current - alpha_right_baseline))

def function2(alpha_left_current, alpha_right_current, alpha_left_baseline, alpha_right_baseline):
    return ((alpha_left_current - alpha_right_current) / (alpha_left_current + alpha_right_current)) - ((alpha_left_baseline - alpha_right_baseline) / (alpha_left_baseline + alpha_right_baseline))

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

# Make data.
X = np.arange(0, 10, 0.2)
Y = np.arange(0, 10, 0.2)
X, Y = np.meshgrid(X, Y)
Z1 = function1(X, Y, 1, 3)
Z2 = function2(X, Y, 1, 3)

# Plot the surface.

surf = ax.plot_surface(X, Y, Z1, cmap=cm.coolwarm,
                       linewidth=0, antialiased=True)
"""
surf = ax.plot_surface(X, Y, Z2, cmap=cm.viridis,
                          linewidth=0, antialiased=True)
"""
ax.set_xlabel('Alpha Left Current')
ax.set_ylabel('Alpha Right Current')
plt.show()
```
