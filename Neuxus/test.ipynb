```python
from mne_lsl.stream import StreamLSL as Stream


stream = Stream(bufsize=60, name='cwl').connect()
stream.info

stream_raw = Stream(bufsize=60, name='raw').connect()
stream_raw.info
```

```python
import matplotlib.pyplot as plt
import numpy as np

data, ts = stream.get_data()
data_raw, ts_raw  = stream_raw.get_data()
plt.figure()
plt.scatter(np.arange(len(ts)), ts)
plt.show()
```

```python
import mne

data, ts = stream.get_data()
data_raw, ts_raw  = stream_raw.get_data()

raw = mne.io.RawArray(data_raw, stream.info)
raw.set_channel_types({'CWL1': 'misc', 'CWL2': 'misc', 'CWL3': 'misc', 'CWL4': 'misc', 'ECG': 'ecg'})
raw.plot_psd(fmax=20);

raw = mne.io.RawArray(data, stream.info)
raw.set_channel_types({'CWL1': 'misc', 'CWL2': 'misc', 'CWL3': 'misc', 'CWL4': 'misc', 'ECG': 'ecg'})
raw.plot_psd(fmax=20);
```
